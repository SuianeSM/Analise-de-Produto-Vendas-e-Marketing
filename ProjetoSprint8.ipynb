## 2. Vendas

Quando as pessoas começam a comprar?

````python
first_touch = df.groupby('Uid', as_index=False)['Start Ts'].min().rename(columns={'Start Ts': 'first_touch'})

first_buy = df2.groupby('Uid', as_index=False)['Buy Ts'].min().rename(columns={'Buy Ts': 'first_buy'})

conversion = pd.merge(first_touch, first_buy, on='Uid', how='inner')

conversion['conversion_days'] = (conversion['first_buy'] - conversion['first_touch']).dt.days

conversion['conversion_category'] = conversion['conversion_days'].apply(
    lambda x: f'{x}d' if x >= 0 else 'erro'
)


conversion = pd.merge(conversion, df[['Uid', 'Source Id']], on='Uid', how='left')
canal_perf = conversion.groupby('Source Id')['conversion_days'].mean().reset_index()

print(conversion.head())
print(conversion['conversion_category'].value_counts(normalize=True))
print('\n Qtde de conversões por canal')
print(canal_perf.sort_values('conversion_days'))
````

````python
media_conversão=conversion['conversion_days'].mean()
print(media_conversão)
````

Quantos pedidos os clientes fazem durante um determinado período de tempo?

````python
df2['year_month'] = df2['Buy Ts'].dt.to_period('M')
first_orders= df2.groupby('Uid').agg({'Buy Ts':'min'})
print(first_orders)
````

````python
first_orders.columns = ['first_order_ts']
first_orders['first_order_dt'] = first_orders['first_order_ts'].dt.date
first_orders['first_order_month'] = first_orders['first_order_ts'].dt.to_period('M')
````

````python
first_orders=first_orders.reset_index()
print(first_orders)
````

````python
coort_users=first_orders.groupby('first_order_month').agg({'Uid':'nunique'}).reset_index()
coort_users.rename(columns={'Uid':'NBuyers'},inplace=True)
print(coort_users)
````

````python
df2['first_order_month'] = (
    df2.groupby('Uid')['year_month']
          .transform('min')          # primeiro mês de compra por uid
)
print(df2)
````

````python

coorts=df2.groupby(['first_order_month','year_month']).size().reset_index(name='NOrders')
print(coorts)
````

````python

coorts['year_month'] = coorts['year_month'].astype('period[M]')
coorts['first_order_month'] = coorts['first_order_month'].astype('period[M]')

coorts['idade_conta'] = (coorts['year_month'] - coorts['first_order_month']).apply(lambda x: x.n)

print(coorts)
````

````python


orders_matrix = (
    coorts.pivot_table(
        index='first_order_month',
        columns='idade_conta',
        values='NOrders',
        aggfunc='sum',
        fill_value=0
    )
).sort_index()

print(orders_matrix)

````

````python
report_coorts=pd.merge(coort_users,coorts,on='first_order_month')
report_coorts['orders_per_buyer']= report_coorts['NOrders']/report_coorts['NBuyers']
````

````python

report_matrix = (
    report_coorts.pivot_table(
        index='first_order_month',
        columns='idade_conta',
        values='orders_per_buyer',
        aggfunc='sum',
        fill_value=0
    )
).sort_index()

print(report_matrix.round(2))
````

````python
plt.figure(figsize=(10, 6))
sns.heatmap(report_matrix, 
            annot=True,
            fmt=".2f",           
            cmap="YlOrBr",       
            linewidths=0.5, 
            cbar_kws={'label': 'Orders per Buyer'})

plt.title("Cohort Analysis - Pedidos por Comprador", fontsize=14)
plt.xlabel("Idade da Conta (meses)")
plt.ylabel("Mês do Primeiro Pedido")
plt.tight_layout()
plt.show()
````

Qual é o volume médio de uma compra?

````python
volume_médio=df2['Revenue'].mean()
print(volume_médio)
````

Quanto dinheiro eles trazem para a empresa (LTV)?

````python
coort_ltv=df2.groupby(['first_order_month','year_month'])['Revenue'].sum().reset_index(name='Valor')
coort_ltv['year_month'] = coort_ltv['year_month'].astype('period[M]')
coort_ltv['first_order_month'] = coort_ltv['first_order_month'].astype('period[M]')

coort_ltv['idade_conta'] = (coort_ltv['year_month'] - coort_ltv['first_order_month']).apply(lambda x: x.n)

report_ltv=pd.merge(coort_users,coort_ltv,on='first_order_month')
report_ltv['ltv']= report_ltv['Valor']/report_ltv['NBuyers']

report_matrix_ltv = (
    report_ltv.pivot_table(
        index='first_order_month',
        columns='idade_conta',
        values='ltv',
        aggfunc='sum',
        fill_value=0
    )
).sort_index()

print(report_matrix_ltv.round(2))
````

````python
plt.figure(figsize=(10, 6))
sns.heatmap(report_matrix_ltv, 
            annot=True,          
            fmt=".2f",          
            cmap="YlGnBu",       
            linewidths=0.5,      
            cbar_kws={'label': 'LTV'}) 

plt.title("Cohort Analysis - LTV por idade da conta", fontsize=14)
plt.xlabel("Idade da Conta (meses)")
plt.ylabel("Mês do Primeiro Pedido")
plt.tight_layout()
plt.show()
````

Quais origens/plataformas você recomendaria? Fundamente sua escolha: em quais métricas você se concentrou? Por quê? Que conclusões você tirou ao encontrar os valores das métricas?

Minha recomendação é manter e escalar o Canal 1, pois ele apresenta o melhor ROI médio (~1,7) com um CAC competitivo, ou seja, cada real investido gera retorno expressivo e com eficiência de aquisição.

O Canal 2, embora caro (CAC alto), mostra um ROI acima de 1 — o que indica que há potencial se forem feitos ajustes de otimização, como melhorar a taxa de conversão ou reduzir o custo de mídia.

O Canal 3, por outro lado, deve ser descontinuado ou revisto, já que apresenta o maior CAC e o pior ROI (< 0,5) — ele destrói margem e não traz retorno financeiro.

Os Canais 4, 5, 9 e 10 se mostram promissores, com CACs baixos e ROI próximos ou ligeiramente acima de 1. Antes de escalar, eu olharia o LTV por origem para confirmar se o valor de vida do cliente justifica o investimento — se o LTV sustentar, são canais eficientes.

**Fundamentação**

As métricas-chave foram:

- ROI (Return on Investment) → mede o retorno de cada real investido por canal.

- CAC (Custo de Aquisição de Cliente) → indica o custo médio para trazer um cliente.

- ROMI por idade da conta → mostra a velocidade de payback (em quantos meses o investimento retorna).

Essas métricas permitem ver eficiência e tempo de retorno, que são mais relevantes do que volume de vendas isolado.

No gráfico de CAC x ROI, o ideal é estar no quadrante superior esquerdo (alto ROI, baixo CAC) — é onde está o Canal 1.
O heatmap de ROMI por idade da conta mostra que, ao longo de 2018, o retorno ficou consistentemente abaixo de 1, indicando queda na eficiência das campanhas — o payback está mais lento, e a empresa precisa redirecionar budget para canais de melhor desempenho.

**Conclusão**

- Escalar: Canal 1, e possivelmente 4, 5, 9 e 10 (se LTV confirmar).

- Otimizar: Canal 2.

- Cortar: Canal 3.

- Revisar estratégia 2018: ROI e ROMI baixos indicam necessidade de reavaliação de mix de canais e mensagens.
